# Below is the 2nd Method which is fast and feasible to use

```python
jobTitle = sal_df.groupby('JobTitle') # it will group all the job titles
```

```python
jobTitle.mean() # mean of all numeric columns
mean_tp = jobTitle.mean()['TotalPay'] # get only total pay mean in a variable
```

```python
mean_tp['DEPUTY DIRECTOR OF INVESTMENTS'] # get a mean for any job title
```

```python
def fillStatus(row): # function with one parameter as row
    JT = row[2] # from row using indexing get the job title
    Totatpay = row[7] # from row using indexing get the Total pay
    TotalPay_mean = mean_tp[JT] # get the mean for jobtitle 'JT'
    if (Totatpay > TotalPay_mean): # condition
        return "FT"
    else:
        return "PT"

NaN_Status_df.iloc[:110531,-1] = NaN_Status_df.iloc[:110531,].apply(fillStatus, axis = 1)
```
